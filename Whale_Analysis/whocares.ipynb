# Data Cleaning

In this section, you will need to read the CSV files into DataFrames and perform any necessary data cleaning steps. After cleaning, combine all DataFrames into a single DataFrame.

Files:

## Whale Returns

Read the Whale Portfolio daily returns and clean the data

```python

# Reading whale returns
whale_returns_csv = Path("Resources/whale_returns.csv")
whale_return_df = pd.read_csv(whale_returns_csv, index_col="Date", infer_datetime_format=True, parse_dates=True)
whale_return_df.head()
```

```python
# Count nulls
whale_return_df.isnull().sum()
```

```python
# Drop nulls
whale_return_df = whale_return_df.dropna().copy()
whale_return_df.isnull().sum()
```

## Algorithmic Daily Returns

Read the algorithmic daily returns and clean the data

```python
# Reading algorithmic returns
algo_returns_csv = Path("Resources/algo_returns.csv")
algo_returns_df = pd.read_csv(algo_returns_csv, index_col="Date", infer_datetime_format=True, parse_dates=True)
algo_returns_df.head()
```

```python
# Count nulls
algo_returns_df.isnull().sum()
```

```python
# Drop nulls
algo_returns_df = algo_returns_df.dropna().copy()
algo_returns_df.isnull().sum()
```

## S&P 500 Returns

Read the S&P500 Historic Closing Prices and create a new daily returns DataFrame from the data. 

```python
# Reading S&P 500 Closing Prices
sp500_history_csv = Path("Resources/sp500_history.csv")
sp500_history_df = pd.read_csv(sp500_history_csv, index_col="Date", infer_datetime_format=True, parse_dates=True)
sp500_history_df.sort_index(inplace=True)
sp500_history_df.head()
```

```python
# Check Data Types
sp500_history_df.dtypes
```

```python
# Fix Data Types
#sp500_history_df.iloc[:,0] = sp500_history_df.iloc[:,0].str.replace('$', '')
sp500_history_df['Close'] = sp500_history_df['Close'].str.replace('$', '')
sp500_history_df['Close'] = sp500_history_df['Close'].astype('float')
sp500_history_df.tail().dtypes
```

```python
# Calculate Daily Returns
sp500_daily_returns = sp500_history_df.pct_change()
sp500_daily_returns.head()
```

```python
# Drop nulls
sp500_daily_returns = sp500_daily_returns.dropna().copy()
sp500_daily_returns.head()
```

```python
# Rename Column
sp500_daily_returns = sp500_daily_returns.rename(columns={"Close": "S&P 500"})
sp500_daily_returns.head()
```

## Combine Whale, Algorithmic, and S&P 500 Returns

```python
# Concatenate all DataFrames into a single DataFrame
joined_daily_returns = pd.concat([whale_return_df, algo_returns_df, sp500_daily_returns], axis='columns', join='inner')
joined_daily_returns.head()
```

## Rolling Statistics

Risk changes over time. Analyze the rolling statistics for Risk and Beta. 

1. Plot the rolling standard deviation of the various portfolios along with the rolling standard deviation of the S&P 500 (consider a 21 day window). Does the risk increase for each of the portfolios at the same time risk increases in the S&P?
2. Construct a correlation table for the algorithmic, whale, and S&P 500 returns. Which returns most closely mimic the S&P?
3. Choose one portfolio and plot a rolling beta between that portfolio's returns and S&P 500 returns. Does the portfolio seem sensitive to movements in the S&P 500?
4. An alternative way to calculate a rolling window is to take the exponentially weighted moving average. This is like a moving window average, but it assigns greater importance to more recent observations. Try calculating the ewm with a 21 day half-life.

```python
# Calculate and plot the rolling standard deviation for
# the S&P 500 and whale portfolios using a 21 trading day window
joined_rolling_std = joined_daily_returns.rolling(window=21).std()
joined_rolling_std.plot(title="21 Day Rolling Standard Deviation", figsize=(20,10))
```

```python
# Construct a correlation table
correlation = joined_daily_returns.corr()
correlation
```

```python
# Calculate Beta for a single portfolio compared to the total market (S&P 500)
# (Your graph may differ, dependent upon which portfolio you are comparing)
rolling_brk_covariance = joined_daily_returns['BERKSHIRE HATHAWAY INC'].rolling(window=30).cov(joined_daily_returns['S&P 500'])
rolling_variance = joined_daily_returns['S&P 500'].rolling(window=30).var()
rolling_brk_beta = rolling_brk_covariance / rolling_variance
rolling_brk_beta.plot(figsize=(20, 10), title='Berkshire Hathaway Inc Beta')
```

```python
# Calculate a rolling window using the exponentially weighted moving average. 
rolling_sos_covariance = joined_daily_returns['SOROS FUND MANAGEMENT LLC'].rolling(window=30).cov(joined_daily_returns['S&P 500'])
#rolling_pls_covariance = joined_daily_returns["PAULSON & CO.INC."].rolling(window=30).cov(joined_daily_returns['S&P 500'])
rolling_tig_covariance = joined_daily_returns['TIGER GLOBAL MANAGEMENT LLC'].rolling(window=30).cov(joined_daily_returns['S&P 500'])
rolling_brk_covariance = joined_daily_returns['BERKSHIRE HATHAWAY INC'].rolling(window=30).cov(joined_daily_returns['S&P 500'])
rolling_alg1_covariance = joined_daily_returns['Algo 1'].rolling(window=30).cov(joined_daily_returns['S&P 500'])
rolling_alg2_covariance = joined_daily_returns['Algo 2'].rolling(window=30).cov(joined_daily_returns['S&P 500'])

rolling_variance = joined_daily_returns['S&P 500'].rolling(window=30).var()

rolling_sos_beta = rolling_sos_covariance / rolling_variance
#rolling_pls_beta = rolling_pls_covariance / rolling_variance
rolling_tig_beta = rolling_tig_covariance / rolling_variance
rolling_brk_beta = rolling_brk_covariance / rolling_variance
rolling_alg1_beta = rolling_alg1_covariance / rolling_variance
rolling_alg2_beta = rolling_alg2_covariance / rolling_variance


ax = rolling_sos_beta.plot(figsize=(20, 10), title='Exponentially Weighted Average')
#rolling_pls_beta.plot(ax=ax)
rolling_tig_beta.plot(ax=ax)
rolling_brk_beta.plot(ax=ax)
rolling_alg1_beta.plot(ax=ax)
rolling_alg2_beta.plot(ax=ax)

ax.legend(["SOROS FUND MANAGEMENT LLC", "TIGER GLOBAL MANAGEMENT LLC", "BERKSHIRE HATHAWAY INC", "Algo 1", "Algo 2"])
```
